# Skewness and Kurtosis

```python
df[['Market_Cap', 'Price']].describe()
```

```python
skewness_market_cap = df['Market_Cap'].skew()
kurtosis_market_cap = df['Market_Cap'].kurtosis()
skewness_price = df['Price'].skew()
kurtosis_price = df['Price'].kurtosis()

print(f'Market Cap - Skewness: {skewness_market_cap}, Kurtosis: {kurtosis_market_cap}')
print(f'Price - Skewness: {skewness_price}, Kurtosis: {kurtosis_price}')
```

```python
sns.histplot(df['Market_Cap'], kde=True)
plt.title('Original Distribution of Market Cap')
plt.show()

sns.histplot(df['Price'], kde=True)
plt.title('Original Distribution of Price')
plt.show()
```

```python
# Apply log transformation to Market Cap and Price
df['Log_Market_Cap'] = np.log(df['Market_Cap'] + 1)  # Adding 1 to avoid log(0)
df['Log_Price'] = np.log(df['Price'] + 1)
```

```python
# Recalculate skewness and kurtosis for the transformed variables
skewness_log_market_cap = df['Log_Market_Cap'].skew()
kurtosis_log_market_cap = df['Log_Market_Cap'].kurtosis()
skewness_log_price = df['Log_Price'].skew()
kurtosis_log_price = df['Log_Price'].kurtosis()

print(f'Log-Transformed Market Cap - Skewness: {skewness_log_market_cap}, Kurtosis: {kurtosis_log_market_cap}')
print(f'Log-Transformed Price - Skewness: {skewness_log_price}, Kurtosis: {kurtosis_log_price}')
```

```python
# Histograms and density plots for the log-transformed Market Cap and Price
sns.histplot(df['Log_Market_Cap'], kde=True)
plt.title('Log-Transformed Distribution of Market Cap')
plt.show()

sns.histplot(df['Log_Price'], kde=True)
plt.title('Log-Transformed Distribution of Price')
plt.show()
```

```python
# Box plots for Market Cap and Price before and after transformation
sns.boxplot(y=df['Market_Cap'])
plt.title('Box Plot of Market Cap')
plt.show()

sns.boxplot(y=df['Log_Market_Cap'])
plt.title('Box Plot of Log-Transformed Market Cap')
plt.show()

sns.boxplot(y=df['Price'])
plt.title('Box Plot of Price')
plt.show()

sns.boxplot(y=df['Log_Price'])
plt.title('Box Plot of Log-Transformed Price')
plt.show()
```

```python
# Box plots for Market Cap and Price by Sector to see if certain sectors have more pronounced skewness or outliers
sns.boxplot(x='Sector', y='Market_Cap', data=df)
plt.title('Market Cap by Sector')
plt.xticks(rotation=90)
plt.show()

sns.boxplot(x='Sector', y='Price', data=df)
plt.title('Price by Sector')
plt.xticks(rotation=90)
plt.show()
```

# Different Visualisation for the Top 10 Sectors

```python
top_sectors = df['Sector'].value_counts().head(10).index
df_top_sectors = df[df['Sector'].isin(top_sectors)]
```

```python
# Set up the figure and axes for the box plots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Price Box Plot
sns.boxplot(ax=axes[0, 0], x='Sector', y='Price', data=df_top_sectors)
axes[0, 0].set_title('Price by Sector')
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=45, ha='right')

# Price/Earnings Box Plot
sns.boxplot(ax=axes[0, 1], x='Sector', y='Price/Earnings', data=df_top_sectors)
axes[0, 1].set_title('Price/Earnings by Sector')
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=45, ha='right')

# Market Cap Box Plot
sns.boxplot(ax=axes[1, 0], x='Sector', y='Market_Cap', data=df_top_sectors)
axes[1, 0].set_title('Market Cap by Sector')
axes[1, 0].set_xticklabels(axes[1, 0].get_xticklabels(), rotation=45, ha='right')

# Dividend Yield Box Plot
sns.boxplot(ax=axes[1, 1], x='Sector', y='Dividend_Yield', data=df_top_sectors)
axes[1, 1].set_title('Dividend Yield by Sector')
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right')

# make it more readable
plt.tight_layout()

# Show plot
plt.show()
```

```python
sns.scatterplot(x='Market_Cap',
                y='Price',
                hue='Sector',data=df_top_sectors,)
                
plt.title('Price vs. Market Cap by Sector (Top 10 Sectors Only)')
plt.xlabel('Market Cap')
plt.ylabel('Price')
plt.show()

```
